```python
import pandas as pd
import random
import string
import numpy as np
from tqdm import tqdm
```

```python
#Importing data
d_data = pd.read_csv("diabetic_data.csv")
d_data
```

```python
#Column names
d_data.columns
```

```python
#Question 1: Melt
#Medications combined into one column titled "Medications"
melted_d_data = d_data.melt(id_vars = ["encounter_id","patient_nbr","race","gender","age","weight","admission_type_id",
                                       "discharge_disposition_id","admission_source_id","time_in_hospital","payer_code",
                                       "medical_specialty","num_lab_procedures","num_procedures","num_medications","number_outpatient","number_emergency",
                                       "number_inpatient","diag_1","diag_2","diag_3","number_diagnoses","max_glu_serum","A1Cresult","change",
                                       "diabetesMed","readmitted"], var_name = "Medications", value_name = "Yes/No")
melted_d_data                            
```

```python
#Question 2: Pivot
#Switching the melted data back with a pivot (medications back into individual
#columns) 
pivot_d_data = melted_d_data.pivot(index = ["encounter_id","patient_nbr","race","gender","age","weight","admission_type_id",
                                       "discharge_disposition_id","admission_source_id","time_in_hospital","payer_code",
                                       "medical_specialty","num_lab_procedures","num_procedures","num_medications","number_outpatient","number_emergency",
                                       "number_inpatient","diag_1","diag_2","diag_3","number_diagnoses","max_glu_serum","A1Cresult","change",
                                       "diabetesMed","readmitted"], columns = "Medications", values = "Yes/No")
pivot_d_data
```

```python
#Question 3: Aggregation
print("Min lab procedures")
d_data["num_lab_procedures"].min()
```

```python
print("Max lab procedures")
d_data["num_lab_procedures"].max()
```

```python
print("Mean lab procedures")
d_data["num_lab_procedures"].mean()
```

```python
print("Standard deviation of number of lab procedures")
d_data["num_lab_procedures"].std()
```

```python
d_data.groupby('gender').agg({
    'num_lab_procedures': ['min','max','mean', 'std'],
    'num_medications': ['min','max','mean', 'std']
})
```

```python
#Question 4: Iteration
#The number of outpatient, emergency, and inpatient encounters are combined
#and added into a new column labeled "total_encounters"

num_encounters = []
for index, row in d_data.iterrows():
    num_encounters.append(row["number_outpatient"] + row["number_emergency"] +
                         row["number_inpatient"])
d_data["total_encounters"] = num_encounters
    
d_data
```

```python
#Question 5: Groupby 

print("Average nuber of lab procedures by gender")
d_data.groupby("gender")["num_lab_procedures"].mean()
```
